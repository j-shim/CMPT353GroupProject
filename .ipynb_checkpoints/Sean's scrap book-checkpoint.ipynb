```python
import sys
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import mstats
import difflib
from datetime import datetime
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import FunctionTransformer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.svm import SVC

wikidata = pd.read_json("data/wikidata-movies.json.gz", lines=True)
rotten = pd.read_json("data/rotten-tomatoes.json.gz", lines=True)
imdb = pd.read_json("data/omdb-data.json.gz", lines=True)
```

```python
#select neccessary columns and remove all rows containing any NaN values
wikidata = wikidata[['cast_member', 'director', 'rotten_tomatoes_id', 'country_of_origin', 'original_language', 'genre', 'publication_date']];
```

```python
# Since highest number of movies are published on thursday and friday, arrange weekdays so that x and y axes have a linear relationship.
def findDtOne(d):
    dtobj = datetime.strptime(d, '%Y-%m-%d')
    day = dtobj.weekday()
    day += 2
    if day == 7:
        day = 0
    elif day == 8:
        day = 1
    return day
findDt = np.vectorize(findDtOne)
wikidata = wikidata.dropna(axis=0, how='any')
wikidata['day'] = findDt(wikidata['publication_date'])
wikidata
plt.hist(wikidata['day'])
```

```python
#select neccessary columns and remove all rows containing any NaN values
rotten = rotten[['audience_average', 'audience_ratings', 'critic_average', 'critic_percent', 'rotten_tomatoes_id']];
rotten = rotten.dropna(axis=0, how='any');
rotten
```

```python
#join all the data tables and remove NaN rows
wr = wikidata.set_index('rotten_tomatoes_id').join(rotten.set_index('rotten_tomatoes_id'))
wr = wr.dropna(axis=0, how='any');
```

```python
# remove index columns and retain only the useful columns.
wr = wr.reset_index()
wr = wr.drop(['rotten_tomatoes_id'], axis=1)
wr
```

## What to do with categorical variables?

The general idea is to convert cast_member, director, and genres columns to meaningful variables before running them on machine learning algorithms.

Cast_member column will be converted to the following features:
count casts who appeared in n>=15 movies, ..in 15>n>=8 movies, ..in 8>n>=1 movies.

Same idea applies for director column.

genres will be converted to dummy variables (basically, add new column for each genre and the value for each column will be boolean. e.g. 'horror' column has 0 or 1 for each movie). Since adding too many features will result in overfitting, we should bag the lowest-occurence genres together into one feature)
(resource: https://blog.myyellowroad.com/using-categorical-data-in-machine-learning-with-python-from-dummy-variables-to-deep-category-66041f734512)

```python
# define count function. count each item in a column and put the count information in a dictionary.
def countEach(listdata, counts):
    for item in listdata:
        counts[item] = counts.get(item, 0) + 1

countCol = np.vectorize(countEach)        

# print the dict keys and values, sorted by its values.
def printSortedDict(dic):
    for w in sorted(dic, key=dic.get, reverse=True):
        print (str(w) + ": " + str(dic[w]))

# plot the dict keys and values, sorted by its values.
def plotSortedDict(dic):
    y_pos = np.arange(len(dic))
    y = []
    for w in sorted(dic, key=dic.get, reverse=True):
        y.append(dic[w])
    plt.bar(y_pos, y, align='center', alpha=0.5)
    plt.tick_params(
        axis='x',          # changes apply to the x-axis
        which='both',      # both major and minor ticks are affected
        bottom=False,      # ticks along the bottom edge are off
        labelbottom=False) # labels along the bottom edge are off
    plt.ylabel('Count')
    plt.figure(figsize=(35,10))
    plt.show()
```

```python
# process the cast_member column
castCounts = {}
casts = np.array(wr['cast_member'])
countCol(casts, castCounts) #outputs count results to castCounts dictionary

#let's get a feel for how many movies each cast has been in (within the list).
printSortedDict(castCounts)
```

```python
#iterate through all the casts of all movies and create a 2D np array which bags them into 4 groups
firstRow = np.zeros((1,3))
for i in range(casts.size):
    newRow = np.zeros((1,3))
    for c in casts[i]:
        if castCounts[c] >= 20:
            newRow[0][0] += 1
        elif castCounts[c] >= 8:
            newRow[0][1] += 1
        elif castCounts[c] >= 1:
            newRow[0][2] += 1
    firstRow = np.concatenate((firstRow, newRow), axis=0)

baggedCasts = np.delete(firstRow, (0), axis=0)
baggedCasts
```

```python
# process the director column
dirCounts = {}
directors = np.array(wr['director'])
countCol(directors, dirCounts) #outputs count results to dirCounts dictionary

#let's get a feel for how many movies each director has directed (within the list).
printSortedDict(dirCounts)
```

```python
#iterate through all the directors of all movies and create a 2D np array which bags them into 3 groups
firstRow = np.zeros((1,3))
for i in range(directors.size):
	newRow = np.zeros((1,3))
	for c in directors[i]:
		if dirCounts[c] >= 10:
		    newRow[0][0] += 1
		elif dirCounts[c] < 10 and dirCounts[c] >= 4:
		    newRow[0][1] += 1
		elif dirCounts[c] < 4 and dirCounts[c] >= 1:
		    newRow[0][2] += 1
	firstRow = np.concatenate((firstRow, newRow), axis=0)

baggedDirectors = np.delete(firstRow, (0), axis=0)
baggedDirectors
```

```python
# process the country_of_origin column
countryCounts = {}
countries = np.array(wr['country_of_origin'])

def countEachCountry(item, counts):
    counts[item] = counts.get(item, 0) + 1

countColCountry = np.vectorize(countEachCountry)     
countColCountry(countries, countryCounts) #outputs count results to countryCounts dictionary
#let's get a feel for how many movies belong in the same countries.
printSortedDict(countryCounts)
```

```python
#iterate through all rows of column countries and create a 2D np array which bags them into 4 groups
firstRow = np.zeros((1,2))
for c in countries:
    newRow = np.zeros((1,2))
    if countryCounts[c] >= 8000:
        newRow[0][0] += 1
    else:
        newRow[0][1] += 1
    firstRow = np.concatenate((firstRow, newRow), axis=0)

baggedCountries = np.delete(firstRow, (0), axis=0)
baggedCountries
```

```python
# process the genre column
genreCounts = {}
genres = np.array(wr['genre'])
countCol(genres, genreCounts) #outputs count results to dirCounts dictionary

#let's get a feel for how many movies each director has directed (within the list).
printSortedDict(genreCounts)
```

```python
#iterate through all the directors of all movies and create a 2D np array which bags them into 3 groups
firstRow = np.zeros((1,3))
for i in range(genres.size):
	newRow = np.zeros((1,3))
	for c in genres[i]:
		if genreCounts[c] >= 2000:
		    newRow[0][0] += 1
		elif genreCounts[c] < 2000 and genreCounts[c] >= 1000:
		    newRow[0][1] += 1
		elif genreCounts[c] < 1000:
		    newRow[0][2] += 1
	firstRow = np.concatenate((firstRow, newRow), axis=0)

baggedGenres = np.delete(firstRow, (0), axis=0)
baggedGenres
```

```python
#concatenate the bagged data to our wr dataframe.
baggedCastsDf = pd.DataFrame(baggedCasts, columns=['casts_A', 'casts_B', 'casts_C'])
baggedDirectorsDf = pd.DataFrame(baggedDirectors, columns=['dirs_A', 'dirs_B', 'dirs_C'])
baggedCountriesDf = pd.DataFrame(baggedCountries, columns=['countries_A', 'countries_B'])
baggedGenresDf = pd.DataFrame(baggedGenres, columns=['genres_A', 'genres_B', 'genres_C'])
wr2 = pd.concat([wr, baggedCastsDf, baggedDirectorsDf, baggedCountriesDf, baggedGenresDf], ignore_index=False, axis=1)
wr2 = wr2.drop(['director', 'cast_member', 'country_of_origin'], axis=1)
wr2

```

## Predicting Critic Average

```python
# Prepare values. Play around with these categories to find the best fit
X = wr2[['casts_A', 'casts_B', 'casts_C', 'dirs_A', 'dirs_B', 'dirs_C', 'countries_A', 'countries_B', 'genres_A', 'genres_B', 'genres_C', 'day']].values
y = (wr2['critic_average'].values/2).astype('int').astype('str')
y.size
```

```python
# Bayesian model
n = 300
total = 0
array = []
for i in range(n):
    X_train, X_valid, y_train, y_valid = train_test_split(X, y)
    bayes_model = GaussianNB()
    bayes_model.fit(X_train, y_train)
    total += bayes_model.score(X_valid, y_valid)
    array.append(bayes_model.score(X_valid, y_valid))
    
print(total/n)
```

```python
#lets see what the distribution of the scores look like
plt.hist(array)
plt.show()
```

```python
# k-nearest neighbours (scaled)
n_nbs = 15
n = 20
total = 0
array = []
for i in range(n):
    X_train, X_valid, y_train, y_valid = train_test_split(X, y)
    knn_model = make_pipeline(
        MinMaxScaler(),
        KNeighborsClassifier(n_neighbors=n_nbs)
        )
    knn_model.fit(X_train, y_train)
    total += knn_model.score(X_valid, y_valid)
    array.append(knn_model.score(X_valid, y_valid))

print(total/n)
```

```python
#lets see what the distribution of the scores look like
plt.hist(array)
plt.show()
```

```python
# Support Vector machine

n = 2
total = 0
array = []
for i in range(n):
    X_train, X_valid, y_train, y_valid = train_test_split(X, y)
    svc_model = SVC(kernel='rbf', C=3, gamma='scale', decision_function_shape='ovr')
    svc_model.fit(X_train, y_train)
    total += svc_model.score(X_valid, y_valid)
    array.append(svc_model.score(X_valid, y_valid))

print(total/n)
```

## Closing remarks:

1. "audience ratings" was actually harmful to the prediction score in all my experiments.

2. At first, I tried joining all of wiki, rotten, and imdb then performing this same experiment, but it didn't work out so well (I had 40% tops), possibly due to not having enough rows to train the models on (when you take out all the NaNs, the data shrinks quite considerably). There is a tradeoff to be made by joining the tables. Perhaps we can use imputation to salvage some rows.

3. Rounding the y values is not a good idea (as explained above). As opposed to flooring or ceiling, rounding hurts our prediction because 0 to 0.5 gets mapped to 0, and 4.5 to 5 gets mapped to 5, while all other numbers are mapped from a range of 1 - this results in uneven categories.

4. Optimizing: I've tried changing the cutoff points for the features (i.e. how to bag the features) for best possible result. The current setting of 'casts_over15', 'casts_8to14', 'casts_1to8', 'dirs_over5', 'dirs_3to4', 'dirs_1to2' seems to most evenly spread out the counts, and thus yield the best prediction results.
