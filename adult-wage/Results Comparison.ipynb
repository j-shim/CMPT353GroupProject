```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
```

```python
results_transformed = pd.read_csv("results_transformed.csv")
results_raw = pd.read_csv("results_raw.csv")
final_result = pd.merge(results_transformed, results_raw, on='Unnamed: 0', how='outer')
final_result = final_result.rename(columns={'Validation Score_y': 'Before Scaling', 'Validation Score_x': 'After Scaling'})

bar = final_result[['Before Scaling','After Scaling']].plot(kind='bar', legend=True, fontsize=12)
bar.set_xticklabels(['Naive\nBayes', 'k-Nearest\nNeighbours', 'Support\nVector\nMachine', 'Neural\nNetwork', 'Gradient\nBoosted\nTree'])
bar.set_title('How Scaling Numerical Features\nChanged the Validation Score')
bar.set_xlabel('Machine Learning Models')
bar.set_ylabel('Validation Score')
bar.set_ylim([0.60,0.86])
```

```python
bar2 = results_transformed[['Training Score','Validation Score']].plot(kind='bar', color=['red', 'green'], legend=True, fontsize=12)
bar2.set_xticklabels(['Naive\nBayes', 'k-Nearest\nNeighbours', 'Support\nVector\nMachine', 'Neural\nNetwork', 'Gradient\nBoosted\nTree'])
bar2.set_title('Training and Validation Scores of All Models')
bar2.set_xlabel('Machine Learning Models')
bar2.set_ylabel('Score')
bar2.set_ylim([0.60,0.90])
```
